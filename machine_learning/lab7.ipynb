# Разделение на блоки 8х8 для массива изображений

```python
data = []
for k in range(400):
    for i in range (8):
        for j in range(8):
            data_temp = faces.images[k, 8*i:(i+1)*8, j*8:(j + 1) *8]
            data.append(data_temp)
print len(data)
print data[0].shape
#print data
```

# Кластеризация

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=8) 
data_new = []
for v in data:
    data_new.append(np.ravel(v))
k_means.fit(data_new)
#y_pred = k_means.predict((data2[1])[2])
```
