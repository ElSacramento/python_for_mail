```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

## Распределение хи-квадрат
### математические ожидание = к
### дисперсия = 2к

```python
k = 6
rv = sts.chi2(k)

#генерируем выборку 
sample = rv.rvs(size=1000)

#гистрограмма 
plt.hist(sample, normed=True)

#теоретическая плотность распределения
x = np.linspace(0,25,100)
pdf = rv.pdf(x)

#построение
plt.plot(x, pdf, label='theoretical pdf', linewidth=5, alpha=0.5)
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

```python
#генерируем выборки
n = 5
array = [rv.rvs(size=n) for i in range(1000)]
```

```python
#расчет среднеквадратичных
averages = []
for item_number, item in enumerate(array):
    s = 0
    for elem in array[item_number]:
        s += elem
    averages.append(s/n)
    
#плотность нормального распределения
norm_rv = sts.norm(loc=k, scale=math.sqrt(2.0*k/n))
x = np.linspace(0,15,100)
pdf = norm_rv.pdf(x)
```

```python
#построение
plt.hist(averages, normed=True)
plt.plot(x, pdf, label='pdf')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```
